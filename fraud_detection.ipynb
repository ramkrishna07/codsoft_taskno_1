# Credit Card Fraud Detection using Logistic Regression

Loading The Train Dataset

```python
import numpy as np
import pandas as pd
train_df=pd.read_csv("D:\Machine Learning\Codsoft Datsets\Credit_Card_Fraud_Detection/fraudTrain.csv")
train_df.head()
```

Loading The Test Dataset

```python
test_df=pd.read_csv("D:\Machine Learning\Codsoft Datsets\Credit_Card_Fraud_Detection/fraudTest.csv")
test_df.head()
```

```python
# training dataset informations
train_df.info()
```

```python
# testing dataset informations
test_df.info()
```

Data cleaning

```python
new_train_df=train_df.drop(columns=['Unnamed: 0','trans_date_trans_time','merchant','first','last','street','city','state','job','dob','trans_num'],axis=1)
new_train_df.head()
```

```python
new_test_df=test_df.drop(columns=['Unnamed: 0','trans_date_trans_time','merchant','first','last','street','city','state','job','dob','trans_num'],axis=1)
new_test_df.head()
```

```python
# checking the number of missing values in each column
new_train_df.isnull().sum()
```

```python
new_test_df.isnull().sum()
```

Applying Label Encdoing on Gender and Category to convert categorical feature to numerical feature

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
new_train_df['category']=le.fit_transform(new_train_df['category'])
new_train_df['gender']=le.fit_transform(new_train_df['gender'])
new_train_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
new_test_df['category']=le.fit_transform(new_test_df['category'])
new_test_df['gender']=le.fit_transform(new_test_df['gender'])
new_test_df.head()
```

```python
# distribution of legit transactions & fraudulent transactions
new_train_df['is_fraud'].value_counts()
```

Applying Feature Scaling using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
new_train_df.iloc[:,0:11]=sc.fit_transform(new_train_df.iloc[:,0:11])
new_train_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
new_test_df.iloc[:,0:11]=sc.fit_transform(new_test_df.iloc[:,0:11])
new_test_df.head()
```

Train Test Split Data

```python
X_train=new_train_df.drop(columns='is_fraud',axis=1)
X_test=new_test_df.drop(columns='is_fraud',axis=1)
```

```python
y_train=new_train_df['is_fraud']
y_test=new_test_df['is_fraud']
```

LogisticRegression Model Builiding and Training & Testing

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=5000)
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train,y_train)
```

model score

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```
